## Part 2

How can we merge the data from the Covid-19 dataset and the Presidential election dataset?

```python
df_county = pd.read_csv('covid_county_population_usafacts.csv')
df_deaths = pd.read_csv('covid_deaths_usafacts.csv')
df_confirmed = pd.read_csv('covid_confirmed_usafacts.csv')
df_super = pd.merge(df_confirmed, df_deaths, on = ['countyFIPS', 'County Name',
                                                   'State', 'StateFIPS'])
df_super = pd.merge(df_county, df_super, on = ['countyFIPS', 'State'])
df_super.head()
```

This is the code for the super class of the covid-19 data. This has every single piece of data all in one that our group made. This is needed to setup the next merge.

### Merge with Presidential Election Results

Here is the completely merged Covid-19 dataset with the Presidential Election dataset. This includes pretty much every comparison you could ask for relating to the election and where the covid-19 cases occured. 

```python
president = pd.read_csv("president_county_candidate.csv")
df_super['County Name'] = df_super['County Name_x'].str.strip()
newcsv = pd.merge(president, df_super, left_on="county", right_on="County Name_x")

newcsv.head()
```

**The individual variable that connected the datasets was the "county" variable in the presidential data and the "County Name_x" in the Covid-19 dataset.**
