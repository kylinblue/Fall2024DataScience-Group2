## Merging COVID-19 Data with enrichment data

```python
import pandas as pd
df_county = pd.read_csv('covid_county_population_usafacts.csv')
df_deaths = pd.read_csv('covid_deaths_usafacts.csv')
df_confirmed = pd.read_csv('covid_confirmed_usafacts.csv')
df_super = pd.merge(df_confirmed, df_deaths, on = ['countyFIPS', 'County Name', 
                                                   'State', 'StateFIPS'])
df_super = pd.merge(df_county, df_super, on = ['countyFIPS', 'State'])
df_super.to_csv('covid_super.csv', index=False)
```

##### exported results to a file names 'covid_super.csv'

##### next I want only data tied to Florida and I will also 
##### delete the State column since it will also be Florida.
##### Because of duplicates I will also drop County Name_y
##### and rename County Name_x to County Name

```python
fl_super = df_super.loc[df_super['State'] == 'FL']
```

```python
fl_super = fl_super.drop(columns=['State'])
```

```python
fl_super = fl_super.drop(columns=['County Name_y'])
fl_super = fl_super.rename(columns={"County Name_x": "County Name"})
```

```python
fl_super.head()
```

```python
fl_super.to_csv('florida_super.csv', index=False)
```

##### Next we will merge the COVID-19 Dataset with our enrichment data

```python
florida_super_super =  pd.merge(fl_super, fl_income, how = 'outer')
```

```python
florida_super_super.to_csv('florida_super_super.csv', index=False)
```

## Compute COVID-19 Trend for Florida

##### First I will get the data from the last 7 columns/days
##### Columns ending in _x are confirmed cases while those ending in _y are confirmed deaths

```python
flData = pd.read_csv('florida_super_super.csv')
```

```python
deaths = flData.iloc[:,-7:]
deaths.drop(labels=0, axis=0, inplace=True)
deaths.head()
```

##### Using the starting date of the last week of data from the deaths data, I found the
##### last week of data from the new cases data

```python
flData.columns.get_loc("2023-07-17_x")
```

```python
cases = flData.iloc[:,1273:1280]
cases.drop(labels=0,axis=0,inplace=True)
cases.head()
```

```python
dSeries = []
cSeries = []
```

```python
for column in deaths:
    print(deaths.loc[:,column].mean())
    dSeries.append(deaths.loc[:,column].mean())
```

```python
for column in cases:
    print(cases.loc[:,column].mean())
    cSeries.append(cases.loc[:,column].mean())
```

```python
deaths_df= pd.DataFrame(dSeries, index=["Day 1","Day 2","Day 3","Day 4","Day 5","Day 6","Day 7",])
cases_df= pd.DataFrame(cSeries, index=["Day 1","Day 2","Day 3","Day 4","Day 5","Day 6","Day 7",])
```

```python
deaths_df.plot.line()
```

```python
cases_df.plot.line()
```
